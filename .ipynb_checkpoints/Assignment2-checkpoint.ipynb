1. <br>
- $\cos{\theta}x+4y+z = \sqrt{3}$ <br>
  not a linear equation because of nonlinear operation √3 and trignometric function cosθ.
- $cos{x}4y+z = \sqrt{3}$ <br>
  not a linear equation because of  trignometric function cos and a square root, and x and y are not seperated which is not following the form.
- $x^{-1}+7y+z = \sin{(\frac{\pi}{2})}$ <br>
  not a linear equation because x is raised to the power of negative one, trignometric function sin(π/2).
- $x^1+y^1+z^1=0$ <br>
  yes it is a linear equation because x,y,z are raised to the first power, there are no trigonometric functions, square roots, and other nonlinear operations.

2. <br>
- $2x_1+3x_2 =5$ <br>
  $2x_1 = 5 - 3x_2 $ <br>
  $\frac{2x_1}{2} =  \frac{5 - 3x_2}{2}$ <br>
  $x_1 =  \frac{5 - 3x_2}{2}$ <br><br>
  $2(\frac{5 - 3x_2}{2}) + 3x_2 = 5$ <br>
  $5 - 3x_2 + 3x_2 $ <br>
  $5 = 5$ <br><br>
- $4x_1+3x_2+2x_3=1$ <br>
  $4x_1 = 1 - 3x_2 - 2x_3  $ <br>
  $ \frac{4x_1}{4} = \frac{1 - 3x_2 - 2x_3}{4}$ <br>
  $x_1 = \frac{1 - 3x_2 - 2x_3}{4}$ <br><br>
  $4(\frac{1 - 3x_2 - 2x_3}{4}) +3x_2+2x_3=1 $ <br>
  $4(\frac{4 - 12x_2 - 8x_3}{4}) +3x_2+2x_3=1 $ <br>
  $ 1 - 3x_2 - 2x_3 + 3x_2+2x_3 = 1$ <br>
  $ 1=1$ 
  <br><br>
